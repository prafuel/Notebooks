```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/social_network_ads.csv", index_col='User ID')
```

```python
df.head()
```

```python
X = df['Age']
y = df['EstimatedSalary']
```

```python
from sklearn.cluster import KMeans

wccs = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit(df[['Age','EstimatedSalary']])
    wccs.append(km.inertia_)
```

```python
plt.plot(list(range(1, 11)), wccs)
```

```python
km = KMeans(n_clusters=5)
y_means = km.fit_predict(df[['Age','EstimatedSalary']])
```

```python
plt.scatter(X, y, c=y_means)
```

```python
gender = np.where(df['Gender'] == 'Male', 0, 1)
plt.scatter(X, y, c=gender)
```
